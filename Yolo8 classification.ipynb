```python
import random
import cv2 as cv
from ultralytics import YOLO
from matplotlib import pyplot as plt
```

```python
# Using the pre trained yolo model to train and predict on the present images.
model = YOLO('yolov8n-cls.pt')
```

```python
# Since YOLO algorithm does data augmentation as a default, passing the images data directly for training.
# Let us run the model for 20 epochs and find the accuracy of the same.

results = model.train(data='/content/drive/MyDrive/FishDataset', epochs = 10)
```

```python
# Function to get the predictions on test images and to plot random predictions with their labels

def get_rgb_image(image):
    rgb_image = cv.cvtColor(image, cv.COLOR_BGR2RGB)
    return rgb_image


def test_accuracy_yolo(test_dir,plot_random=False, n_samples = 8):
    #test_dir = '/content/drive/MyDrive/FishDataset/test'
    import math

    test_folders = os.listdir(test_dir)
    i = 0
    v = 0
    result_dict = {}

    for folder in test_folders:
        results = model(f'{test_dir}/{folder}', verbose=False)
        result_dict.update({folder:results})
        for result in results:
            i += 1
            top1 = result.probs.top1
            classes = result.names
            top1_class_name = classes[top1]

            if top1_class_name == folder:
                v+=1
            #top5 = result.probs.top5
            #top5_class_name = []
            #[top5_class_name.append(classes[cnum]) for cnum in top5]
            #top5cont = result.probs.top5conf.tolist()

    test_accuracy = v/i

    if plot_random == True:
        c = 4
        r = math.ceil(n_samples/c)
        plt.figure(figsize=(20,5*r+1))
        plt.suptitle(f'Visualizing random values with their labels.\nThe accuracy on the data is passed {str("%.2f" % (test_accuracy*100))}%')
        for i in range(n_samples):

            random_label = random.choice(list(result_dict.keys()))
            random_path = f'{test_dir}/{random_label}'
            pred_vals = result_dict[random_label]
            random_result = random.choice(pred_vals)
            random_top1 = random_result.probs.top1
            classes = random_result.names
            random_top1_class_name = classes[random_top1]
            random_top1cont = random_result.probs.top1conf.tolist()

            plt.subplot(r,c,i+1)
            plt.imshow(get_rgb_image(random_result.orig_img))
            plt.ylabel(f'Actual : {random_label}')
            plt.xlabel(f'Predicted : {random_top1_class_name}')
            plt.title(f'Confidence Interval : {str("%.2f" % random_top1cont)}')

    return test_accuracy, result_dict, classes
```

```python
# Predicting on validation set , getting the accuracy and plotting random predictions

val_dir = '/content/drive/MyDrive/FishDataset/test'
val_accuracy, predict_dict, classes = test_accuracy_yolo(val_dir,plot_random=True, n_samples=16)
print(f'The accuracy on Validation data is {str("%.2f" % (val_accuracy*100))}%')
```

```python
# Let us plot confusion matrix to identify on which of the images is the model missing out.
# Also, just checking the images where the model has missed out and ignornig the perfectly predicted categories.

cm = model.metrics.confusion_matrix.matrix
cm_df = pd.DataFrame(data = cm)
new_df = cm_df.set_axis(cm_df.columns.map(classes), axis=1)

for i in range(len(new_df)):
    if cm_df[i][i] == 5 and cm_df[i].sum() == 5 and new_df.iloc[i].sum() == 5:
        cm_df.drop(i, axis=0, inplace=True)
        cm_df.drop(i, axis=1, inplace=True)
        break


mappers = cm_df.columns.map(classes)
mapped_df = cm_df.set_axis(list(mappers), axis=1)
mapped_df = mapped_df.set_axis(list(mappers), axis=0)

import seaborn as sns
plt.figure(figsize=(10,8))
sns.heatmap(mapped_df, annot=True, cmap='OrRd')
plt.xlabel('Prediciton')
plt.ylabel('Truth')
plt.title('Confusion Matrix')
plt.show()
```
